#Adjust datatypes and filling gthe missing values

```python
#convert object datatype to datetime64
netflix_df["date_added"]=pd.to_datetime(netflix_df["date_added"])
```

```python
netflix_df.head()
```

```python
netflix_df.info()
```

```python
netflix_df.fillna({"director":"Unavailable","cast":"Unavailable","country":"Unavailable","rating":"Unavailable"},inplace=True)
```

```python
netflix_df.isna().sum()
```

"""For nulls in date_added,mising date added is to be substituted with the most recent date from date_added. 
This is because netflix has a tendency to add more content over time
other viable option would be finding actual dates & inputing them manually or droping a data from results since the amount of missing data would be small""""


```python
netflix_df[netflix_df.date_added.isnull()]
```

```python
most_recent_date=netflix_df['date_added'].max()
```

```python
netflix_df.fillna({'date_added':most_recent_date},inplace=True)
```

```python
netflix_df[netflix_df.show_id=='s8183']
```

```python
netflix_df[netflix_df.duration.isnull()]
```

The missing duration are all directed by Louis C.K. Normally we would like to fill the  missing values by the mean duration from the table. 
In this case it appears that actual duration was input into the rating column, so one solution is to move the rating data into the duration column and makes the rating information Unavailable like other nulls

```python
#Make sure there is no other content by the same director to avoid the accidental overwriting
netflix_df[netflix_df.director=='Louis C.K.'].head()
```

```python
#Overwrite and check
netflix_df.loc[netflix_df['director']=='Louis C.K.','duration']=netflix_df['rating']#locs helps us accessing column by their name
netflix_df[netflix_df.director=='Louis C.K.'].head()
```

```python
#Second overwrite and check
netflix_df.loc[netflix_df['director']=='Louis C.K.','rating']='Unavailable'
netflix_df[netflix_df.director=='Louis C.K.'].head()

```

Visualization

```python
#Lets take a look at the type of shows that has been watched on netflix

netflix_df.type.value_counts() #value_count method give a counts of diff categories

```

Country analysis

```python
netflix_df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=netflix_df['country'].value_counts().index[0:10],data=netflix_df)
plt.title('Country wise content on netflix')
```

```python
#Checking type of shows based on country
movie_countires=netflix_df[netflix_df['type']=='Movie']
tvshow_countries=netflix_df[netflix_df['type']=='TV Show']
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=netflix_df['country'].value_counts().index[0:10],data=movie_countires)
plt.title('top 10 countries producing movies in netflix')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='country',order=netflix_df['country'].value_counts().index[0:10],data=tvshow_countries)
plt.title('top 10 countries producing Tv_shows in netflix')
```
